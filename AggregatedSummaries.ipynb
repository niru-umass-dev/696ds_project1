```python
import pandas as pd
import os
datasets = ['similarity', 'selfparaphrase', 'paraphrase', 'negation']
experiments = [
    '20230613_234743_agg-final-01-biascont-aggall'
]

cols = ['data_type', 'mean(ds)', 'std_dev(ds)', 'corr(ds)', 'rank_corr(ds)',
       'mean(bs)', 'std_dev(bs)', 'corr(bs)', 'rank_corr(bs)',
       'mean(nli_contrast_ent)', 'std_dev(nli_contrast_ent)',
       'corr(nli_contrast_ent)', 'rank_corr(nli_contrast_ent)',
       'mean(nli_contrast_neut)', 'std_dev(nli_contrast_neut)',
       'corr(nli_contrast_neut)', 'rank_corr(nli_contrast_neut)',
       'mean(nli_contrast_cont)', 'std_dev(nli_contrast_cont)',
       'corr(nli_contrast_cont)', 'rank_corr(nli_contrast_cont)',
       'mean(nli_contrast_agg)', 'std_dev(nli_contrast_agg)',
       'corr(nli_contrast_agg)', 'rank_corr(nli_contrast_agg)']
keep_cols = [
    'data_type', 
    'mean(ds)',
    'rank_corr(ds)',
    'mean(bs)',
    'rank_corr(bs)',
    'mean(nli_contrast_ent)',
    'rank_corr(nli_contrast_ent)',
    'mean(nli_contrast_neut)',
    'rank_corr(nli_contrast_neut)',
    'mean(nli_contrast_cont)',
    'rank_corr(nli_contrast_cont)',
    'mean(nli_contrast_agg)',
    'rank_corr(nli_contrast_agg)'
]
drop_cols = [col for col in cols if col not in keep_cols]

experiment_frames = []
for experiment in experiments:
    dataset_frames = []
    first = True
    for dataset in datasets:
        filename = f"ref_experiment_{dataset}_summary.csv"
        filepath = os.path.join('/work/pi_adrozdov_umass_edu/pgeorge_umass_edu/696ds_project1/data/results/', experiment, filename)
        dataset_df = pd.read_csv(filepath)
        if not first:
            dataset_df = dataset_df.drop(index=0)
        dataset_df = dataset_df.drop(columns=drop_cols)
        dataset_frames.append(dataset_df)
        first = False
        # display(dataset_df)
    
    experiment_df = pd.concat(dataset_frames)
    experiment_df['experiment_name'] = experiment
    experiment_frames.append(experiment_df)
    display(experiment_df)
    
# display(df.columns)


```
